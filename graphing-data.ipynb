```python
import matplotlib.pyplot as plt
import csv
import numpy as np
```

```python
# GROVERS ALGORITHM

plt.style.use("seaborn")
fig = plt.figure(figsize=(12, 8))

qubits, qpp, pyquil, cirq, qiskit = np.loadtxt('grovers.csv', delimiter=',', unpack=True)


plt.plot(qubits,qpp, marker='o',label='Qunatum++', color='blue', linestyle=':', markersize=10)
plt.plot(qubits,pyquil, marker='o',label='PyQuil', color='orange', linestyle=':', markersize=10)
plt.plot(qubits,cirq, marker='o',label='Cirq', color='red', linestyle=':', markersize=10)
plt.plot(qubits,qiskit, marker='o',label='Qiskit', color='purple', linestyle=':', markersize=10)

plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

plt.xlabel('Number of Qubits', fontsize=20)
plt.ylabel('Time Taken (in seconds)', fontsize=20)
plt.title('Qubits vs. Time. Grovers Algorithm', fontsize=20)
plt.legend(fontsize=20)
#plt.yscale('log')

plt.show()
```
