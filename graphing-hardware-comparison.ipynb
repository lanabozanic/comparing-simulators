```python
#plot hardware comparisons
from pandas import *
import matplotlib.pyplot as plt
import csv
import numpy as np
import random

plt.style.use("seaborn")
fig = plt.figure(figsize=(12, 7))

# IBM -> IBM device
# R -> rigetti device
# Ionq -> Ionq device
# honeywell -> honeywell device
# G -> google device
# QT -> QuTech device

devices = ['IBM Tenerife','IBM Yorktown', 'IBM Vigo', 'IBM Oursense', 'IBM Melbourne', 'IBM Poughkeepsie', 'IBM Tokyo', 'IBM Austin', 'IBM System One', 'IBM Singapore', 'IBM Rochester', ' ','Rigetti Aspen-16', 'Rigetti Aspen-7', 'Rigetti Aspen-8', ' ', 'IonQ 11 Qubits', ' ', 'Honeywell 4 Qubit', ' ', 'Google Sycamore', ' ', 'QuTech Spin 2', 'QuTech Starmon 5']



qubits, connectivity, t1, t2, qfid_1, qfid_2  = np.loadtxt('hardware-data.csv', delimiter=',', unpack=True)
```

```python

# PLOT QUBIT NUMBER
fig = plt.figure(figsize=(12, 7))

color= plt.cm.RdPu(np.linspace(.4, .8, 30))

height = qubits
bars = devices
y_pos = np.arange(len(bars))

plt.bar(y_pos, height, color=color)
plt.xticks(y_pos, bars, rotation='vertical')

plt.xlabel('Device Name', fontsize=20)
plt.ylabel('Number of Qubits', fontsize=20)
plt.title("Comparison of Today's Hardware: Number of Qubits", fontsize=20)
plt.xticks(fontsize=15)
plt.yticks(fontsize=20)
plt.show()

```

```python
# PLOT QUBIT CONNECTIVITY

plt.style.use("seaborn")
fig = plt.figure(figsize=(12, 7))

color= plt.cm.spring(np.linspace(.4, .8, 30))


#print(df[0][10])

height = connectivity
bars = devices
y_pos = np.arange(len(bars))

plt.bar(y_pos, height, color=color)
plt.xticks(y_pos, bars, rotation='vertical')

plt.xlabel('Device Name', fontsize=20)
plt.ylabel('Qubit Connectivity', fontsize=20)
plt.title("Comparison of Today's Hardware: Qubit Connectivity", fontsize=20)
plt.xticks(fontsize=15)
plt.yticks(fontsize=20)
plt.show()

```

```python
delete = [16, 22]
filtered_dev = np.delete(devices, delete)
filtered_t1 = np.delete(t1, delete)
plt.style.use("seaborn")
fig = plt.figure(figsize=(12, 7))

color= plt.cm.cool(np.linspace(.4, .8, 30))

#devices.remove('IonQ 11 Qubits')
#t1.remove(10000000000)

print(len(devices))

#print(df[0][10])

height = filtered_t1
bars = filtered_dev
y_pos = np.arange(len(bars))

plt.bar(y_pos, height, color=color)
plt.xticks(y_pos, bars, rotation='vertical')

plt.xlabel('Device Name', fontsize=20)
plt.ylabel('Relaxation time (in microseconds)', fontsize=20)
plt.title("Comparison of Today's Hardware: Relaxation Time", fontsize=20)
plt.xticks(fontsize=15)
plt.yticks(fontsize=20)
plt.show()
```

```python
# PLOT COHERENCE TIME
delete = [16]
filtered_dev = np.delete(devices, delete)
filtered_t2 = np.delete(t2, delete)
plt.style.use("seaborn")
fig = plt.figure(figsize=(12, 7))

color= plt.cm.winter(np.linspace(.4, .8, 30))

#devices.remove('IonQ 11 Qubits')
#t1.remove(10000000000)

print(len(devices))

#print(df[0][10])

height = filtered_t2
bars = filtered_dev
y_pos = np.arange(len(bars))

plt.bar(y_pos, height, color=color)
plt.xticks(y_pos, bars, rotation='vertical')

plt.xlabel('Device Name', fontsize=20)
plt.ylabel('Coherence Time (in microseconds)', fontsize=20)
plt.title("Comparison of Today's Hardware: Coherence Time", fontsize=20)
plt.xticks(fontsize=15)
plt.yticks(fontsize=20)
plt.show()
```

```python
# PLOT 1 QUBIT GATE FIDELITY

plt.style.use("seaborn")
fig = plt.figure(figsize=(12, 7))

color= plt.cm.Wistia(np.linspace(.4, .8, 30))

print(len(devices))


height = qfid_1
bars = devices
y_pos = np.arange(len(bars))

plt.bar(y_pos, height, color=color)
plt.xticks(y_pos, bars, rotation='vertical')

plt.xlabel('Device Name', fontsize=20)
plt.ylabel('Qubit Fidelty (%)', fontsize=20)
plt.title("Comparison of Today's Hardware: 1 Qubit Gate Fidelity", fontsize=20)
plt.xticks(fontsize=15)
plt.yticks(fontsize=20)
plt.show()
```
